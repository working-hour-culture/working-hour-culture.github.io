# Foursquare Check-in Data Cleaning

## Data Import

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
check = pd.read_csv('C:/Users/Desktop/UCL/Group Mini Project/Assessment/Data/check_ins.csv')
```

```python
# Select only Office Type
check = check.loc[check['venue_type'] == 'Office']
```

```python
# Choose the six representative countries
check_select = check.loc[(check['country_code']== 'GB') | (check['country_code']== 'US') | (check['country_code']== 'BR') 
| (check['country_code']== 'SG') | (check['country_code']== 'AU') | (check['country_code']== 'ZA')]
```

```python
check_select.country_code.value_counts()
```

## Data Cleaning

```python
# In order to show the year clearly later
check_select['datetime'] = check_select['datetime'].str.replace("\+0000 ", "")
```

```python
# Extract weekday
check_select['weekday'] = check_select['datetime'].astype(str).str[:3]
```

```python
# Create a new column and convert to datetime for further analysis
check_select['new_datetime'] = pd.to_datetime(check_select['datetime'])
```

```python
# Extract time only
time = pd.Series(pd.to_datetime(check_select['new_datetime']))
# Round to hour
time = time.dt.round('H')
check_select['Time'] = time
# Extract time only in datetime
check_select['Time'] = pd.to_datetime(check_select['Time'],format= '%H:%M:%S' ).dt.time
```

## Divide Data into Weekday and Weekend and Setting Data Types

```python
# Divide into weekday and weekend
check_weekday = check_select[(check_select['weekday']== 'Mon') | (check_select['weekday']== 'Tue') | (check_select['weekday']== 'Wed')
                             | (check_select['weekday']== 'Thu') | (check_select['weekday']== 'Fri')]
check_weekend = check_select[(check_select['weekday']== 'Sat') | (check_select['weekday']== 'Sun')]
```

```python
# random set a date which used for calculation later
check_weekday['Date'] = '2012-01-01'
check_weekend['Date'] = '2012-01-02'
```

```python
# Weekday
# Convert type to datetime
check_weekday['Date'] = pd.to_datetime(check_weekday['Date'], errors='coerce')
check_weekday["Date"] = [d.date() for d in check_weekday["Date"]]
# Covvert the format of date
check_weekday['Date'] = ['{}-{}-{}'.format(d,m,y) for y, m, d in map(lambda x: str(x).split('-'), check_weekday['Date'])]

# Weekend
# Convert type to datetime
check_weekend['Date'] = pd.to_datetime(check_weekend['Date'], errors='coerce')
check_weekend["Date"] = [d.date() for d in check_weekend["Date"]]
# Covvert the format of date
check_weekend['Date'] = ['{}-{}-{}'.format(d,m,y) for y, m, d in map(lambda x: str(x).split('-'), check_weekend['Date'])]
```

## Unix Timestamp Conversion

```python
# Convert type to string

# Weekday
check_weekday['Date']=check_weekday['Date'].astype(str)
check_weekday['Time']=check_weekday['Time'].astype(str)
# Weekend
check_weekend['Date']=check_weekend['Date'].astype(str)
check_weekend['Time']=check_weekend['Time'].astype(str)
```

```python
dt1 = check_weekday.Date.str.cat(check_weekday.Time, sep = " ")
dt2 = check_weekend.Date.str.cat(check_weekend.Time, sep = " ")
```

```python
from datetime import datetime
import time
import datetime
import os
```

```python
# Convert to unixtime
check_weekday['unix'] = [time.mktime(datetime.datetime.strptime(x, "%d-%m-%Y %H:%M:%S").timetuple()) for x in dt1]
check_weekend['unix'] = [time.mktime(datetime.datetime.strptime(x, "%d-%m-%Y %H:%M:%S").timetuple()) for x in dt2]
```
